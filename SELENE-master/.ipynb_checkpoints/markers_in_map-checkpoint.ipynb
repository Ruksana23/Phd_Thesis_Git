```python
import os
from matplotlib.colors import ListedColormap
import time, datetime
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.basemap import Basemap
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
```

```python
data = xr.open_dataset("precip.mon.mean.nc")
data1= data.mean(dim='time')
precip=data1.precip*365
precip
```

```python
sst = xr.open_dataset("sst.nc")
# sst
anom=sst['anom']
# 
```

```python
interpolated_data = precip.interp(coords={'lat': anom['lat'], 'lon': anom['lon']}, method='linear')
interpolated_data
```

```python
ponts_Sr_least=pd.read_csv("ponts_Sr_least.csv")
ponts_Sr_mod=pd.read_csv("ponts_Sr_mod.csv")
ponts_Sr_high=pd.read_csv("ponts_Sr_high.csv")
ponts_Sr_very_high=pd.read_csv("ponts_Sr_very_high.csv")

lon_least = ponts_Sr_least.Longitude
lat_least = ponts_Sr_least.Latitude

lon_mod = ponts_Sr_mod.Longitude
lat_mod = ponts_Sr_mod.Latitude

lon_high = ponts_Sr_high.Longitude
lat_high = ponts_Sr_high.Latitude

lon_very_high= ponts_Sr_very_high.Longitude
lat_very_high = ponts_Sr_very_high.Latitude


```

```python
blue_colors = ['#2166ac', '#4393c3', '#92c5de', '#d1e5f0']
cmap = ListedColormap(blue_colors)
```

```python
fig = plt.figure(figsize=(18, 8))
# ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
# ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())
# ax.add_feature(cfeature.COASTLINE)


ocean_mask = xr.where(np.isnan(anom), 1, np.nan)
masked_precip = interpolated_data * ocean_mask
masked_precip.plot(cmap=cmap)
cbar = plt.colorbar()
cbar.ax.set_yticks([0, 500, 1000, 1500, 2000])
cbar.ax.set_yticklabels(['0', '500', '1000', '1500', '2000'])

plt.scatter(lon_least, lat_least, s=15, c='gold', label="Least")
plt.scatter(lon_mod,lat_mod,  s=15, c='darkorange', label="Moderate")
plt.scatter(lon_high,lat_high,  s=15, c='red', label="High")
plt.scatter(lon_very_high,lat_very_high, s=15, c='maroon', label="Very high")
plt.title("Global Precipitation")
plt.legend()
plt.savefig("Precipitation_with_marks.png")
```
