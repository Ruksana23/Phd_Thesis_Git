```python
import pandas as pd
import matplotlib.lines as mlines
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os
from scipy.signal import medfilt
import matplotlib.patches as mpatches
import datetime as dt  # Import datetime module

# List of shade periods (start and end dates)
flood_periods = [(dt.datetime(2010, 11, 1), dt.datetime(2010, 12, 30)),
                 (dt.datetime(2012, 11, 1), dt.datetime(2012, 11, 10)),
                 (dt.datetime(2020, 11, 20), dt.datetime(2020, 12, 5))
                 # Add more shade periods as needed
                 ]

dcsc_periods = [
    (dt.datetime(2020, 11, 1), dt.datetime(2020, 11, 30))
    # Add more shade periods as needed
]

# Directory containing station data files
file_path = "datafiles/sure_files/UTide_files/Visakhapatnam_output_UTide1.csv"

# Read the data for each station
df = pd.read_csv(file_path)
df['Time'] = pd.to_datetime(df['Time'])

# Perform linear regression on the original data
slope, intercept, r_value, p_value, std_err = stats.linregress(
    df.index,
    df['sl_anomaly']
)

# Convert the slope to mm per year
slope_mm_per_year = slope * 365 * 24  # Assuming seconds as the time unit

# Create a time series for the linear regression line
sea_level_filtered = medfilt(df['sl_anomaly'], kernel_size=((3 * 24) + 1))
# linear_regression_line = (slope * df.index) + intercept

# Plot the sea level data and the linear regression line
plt.figure(figsize=(12, 6))

for dcsc_start, dcsc_end in dcsc_periods:
    plt.gca().axvspan(dcsc_start, dcsc_end, color='cyan', alpha=0.5)
for flood_start, flood_end in flood_periods:
    plt.gca().axvspan(flood_start, flood_end, color='orange', alpha=0.8)

adt = pd.read_csv('average_adt_Visakhapatnam.csv')

# Convert the 'time' column to a datetime object
adt['time'] = pd.to_datetime(adt['time'])

# Set the 'time' column as the DataFrame index
adt.set_index('time', inplace=True)

# Filter data for time starting from 2011
adt = adt[adt.index >= '2011-01-01']
sl = adt['adt'] * 1000
sl_m = sl - sl.mean()

# Create a time series plot
plt.xlabel('Time')
plt.ylabel('Sea Level (mm)')
plt.title(f'Sea Level Trend for Visakhapatnam ({slope_mm_per_year:.2f} mm/year) ')

flood_legend = mpatches.Patch(color='orange', alpha=0.8, label='Flood Period')
dcsc_legend = mpatches.Patch(color='cyan', alpha=0.5, label='D_C_SC Period')
df['sl_anomaly_ci'] = df['sl_anomaly']

# Replace values outside the confidence interval with NaN

confidence_level = 0.95

# Calculate the standard error of the mean
std_err = stats.sem(df['sl_anomaly'], nan_policy='omit')

# Calculate the margin of error
margin_of_error = std_err * stats.t.ppf((1 + confidence_level) / 2, len(df['sl_anomaly']) - 1)

# Calculate the mean of the 'sl_anomaly' column
mean_sl_anomaly = df['sl_anomaly'].mean()

# Calculate the confidence interval
lower_bound = mean_sl_anomaly - margin_of_error
upper_bound = mean_sl_anomaly + margin_of_error
df.loc[(df['sl_anomaly'] > lower_bound) | (df['sl_anomaly'] < upper_bound), 'sl_anomaly_ci'] = np.nan



ax = plt.gca()
plt.plot(df['Time'], sea_level_filtered, linestyle='-', color='blue', linewidth=0.7, label='3day_med_fil_sla INCOIS')
plt.plot(df['Time'], df['sl_anomaly_ci'], linestyle='-', color='black', linewidth=1.5,
         label='sl_anomaly_ci (within CI)')
# plt.plot(df['Time'], linear_regression_line, linestyle='--', color='black', linewidth=2.5,
#          label=f'Linear Regression trend ({slope_mm_per_year:.2f} mm/year)')
plt.plot(adt.index, sl_m, color='red', linewidth=1.5, label='daily_sla_Satellite Altimetry')

# Set the background color of the axis to a custom color (e.g., light gray)
ax.set_facecolor('lightgray')

legend1 = plt.legend(handles=[flood_legend, dcsc_legend], loc='upper left')

# Add a legend for the trend lines
custom_lines = [mlines.Line2D([], [], color='blue', label='3day_med_fil_sla INCOIS'),
                mlines.Line2D([], [], color='red', label='daily_sla_Satellite Altimetry'),
                mlines.Line2D([], [], color='black',label=f'sl_anomaly_ci (within CI)')
               ]

legend2 = plt.legend(handles=custom_lines, loc='upper right')

# Add both legends to the plot
plt.gca().add_artist(legend1)
plt.gca().add_artist(legend2)

plt.ylim(-1500, 1500)
plt.grid(True, linestyle=':', linewidth=0.8, color='Ivory', alpha=0.5)
plt.savefig(f'Trend_for_Visakhapatnam')
plt.show()
```
