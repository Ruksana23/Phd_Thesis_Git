```python
paths=[]
for i in df.Station:
    directory_path= "datafiles/"+i
    sub_files=os.listdir(directory_path)
    sub_files=natsort.natsorted(sub_files,reverse=False)
    for j in sub_files:
        full_path = os.path.join(directory_path, j)
        paths.append(full_path)
```

```python
for i in paths:
    df = pd.read_csv(i,header=None,usecols=[1,2],names=['time','sl'])
    df['sl']=df['sl']*10
    df['time'] = pd.to_datetime(df['time'])
    df.set_index('time', inplace=True)
    df['sl'] = pd.to_numeric(df['sl'], errors='coerce')
    # Resample the data to 1-minute intervals, filling missing values with NaN
    df_resampled = df.resample('1T').asfreq()
    df_resampled.fillna(-99999999, inplace=True)
    df['sl'].replace(99999.99, -99999999, inplace=True)
    
    df_final = df_resampled.reset_index(drop=False)
    df_final['flag'] = 1
    df_final['slno'] = 1
    df_final['date'] = df_final['time'].dt.strftime('%Y%m%d')
    df_final['time'] = df_final['time'].dt.strftime('%H%M%S')
    # Rearrange columns to match the desired order
    df_transformed = df_final[['slno', 'date', 'time', 'sl', 'flag']]
    df_transformed['sl'] = df_transformed['sl'].astype(np.int64)
    df_transformed = df_transformed.astype(str)
    # Combine 'date' and 'time' into one column and then concatenate all columns into a single string column
    df_transformed['combined'] = df_transformed['slno'] + ' ' + \
                               df_transformed['date'] + ' ' + \
                               df_transformed['time'] + ' ' + \
                               df_transformed['sl'] + ' ' + \
                               df_transformed['flag']
    # The final DataFrame should contain only the 'combined' column
    df1 = df_transformed[['combined']]   
    file_name = i.split('/')[-1].replace('.csv', '')
    df1.to_csv("datafiles/"+file_name+'.data', header=False, index=False)
```
