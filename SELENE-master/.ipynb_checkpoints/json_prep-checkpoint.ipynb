```python
df = pd.read_csv("datafiles/INCOIS_Readme.txt", delimiter=r"\s+")
df=df[:-1]
df.head()
# df1=df.loc[df['Longitude'] >= 77.53]
# df2=df1.sort_values(by='Latitude',ascending=False, ignore_index=True)
# df2
```

```python
paths=[]
for i in df.Station:
    directory_path= "datafiles/"+i
    sub_files=os.listdir(directory_path)
    sub_files=natsort.natsorted(sub_files,reverse=False)
    for j in sub_files:
        full_path = os.path.join(directory_path, j)
        paths.append(full_path)
```

```python
import json

# Assuming df is your DataFrame and it's already defined as in the first screenshot

# Create an empty dictionary to hold the JSON structure
stations_data = {}

for i in paths:
    parts = i.split('/')  # Split the string by the slash character
    station = parts[2] 
    year=i.split('/')[-1].split('_')[-1].replace('.csv', '')
    # Extract station name, latitude, and longitude
    latitude = df.loc[df['Station'] == station, 'Latitude'].values[0]
    longitude = df.loc[df['Station'] == station, 'Longitude'].values[0]
    
    station_year_key = f"{station}_{year}"

    # Create a dictionary for the current station_year
    stations_data[station_year_key] = {
        "foremanharmfile": "",
        "latitude": latitude,
        "longitude": longitude,
        "name": station_year_key,
        "maxlevel": 10000,
        "minlevel": -10000,
        "qc_level_nsigma": 4,
        "qc_level_splinedegree": 2,
        "qc_level_winsize": 200,
        "qc_stucklimit": 10,
        "qc_surge_nsigma": 5,
        "qc_surge_splinedegree": 3,
        "qc_surge_winsize": 500,
        "seriesdatecolumns": "1,2",
        "seriesdateformat": "%Y%m%d%H%M%S",
        "seriesfile": f"datafiles/{station_year_key}.data",
        "seriesqccolumn": 4,
        "seriesseparator": " ",
        "seriesvaluecolumn": 3
    }

# Convert the dictionary to a JSON string
stations_json = json.dumps(stations_data, indent=4)

# Print the JSON string
print(stations_json)

with open('configuration/stations.json', 'w') as file:
    file.write(stations_json)
```
