```python
paths=[]
stations=[]
directory_path= "outputs"
sub_files=os.listdir(directory_path)
sub_files=natsort.natsorted(sub_files,reverse=False)
for j in sub_files:
    if '5min' in j:
        full_path = os.path.join(directory_path, j)
        parts = full_path.split('/')  # Split the string by the slash character
        station_year = parts[-1].replace('_5min_interpolated.out', '')
            
        # stations.append(station_year)
        paths.append(full_path)
```

```python
paths
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Define the maximum gap to fill
max_gap = 5

# Process each file in the paths array
for file_path in paths:
    df = pd.read_csv(file_path, names=["Time", "Sl"], delimiter=r"\s+")
    df['Sl'].replace("None", np.nan, inplace=True)
    df['Sl'] = df['Sl'].astype(float)
    
    # Set 'Time' as the index
    df["Time"] = pd.to_datetime(df["Time"])
    df = df.set_index(['Time'])
    
    # Resample to 1-minute frequency
    df = df.resample("1T").asfreq()
    
    # Interpolate gaps that are less than or equal to 5 minutes
    df['Sl'] = df['Sl'].interpolate(method='time', limit=max_gap)
    # Reset index before plotting
    df.reset_index(inplace=True)

    # Save the modified DataFrame to a new CSV file
    output_file_path = file_path.replace('.out', '.csv')  # Create a new file name
    df.to_csv(output_file_path, index=False)
```
