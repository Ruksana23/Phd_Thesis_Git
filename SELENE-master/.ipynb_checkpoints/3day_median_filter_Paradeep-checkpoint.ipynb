```python
import pandas as pd
import matplotlib.lines as mlines
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os
from scipy.signal import medfilt
import matplotlib.patches as mpatches
import datetime as dt  # Import datetime module


```

```python
from scipy.stats import linregress
from scipy import interpolate
file_path = "datafiles/sure_files/UTide_files/Tuticorin_output_UTide.csv"
# Read the data for each station
df = pd.read_csv(file_path)
df['Time'] = pd.to_datetime(df['Time'])
flood_periods = [(dt.datetime(2010, 11, 1), dt.datetime(2010, 12, 30)), 

              		 (dt.datetime(2012, 11, 1), dt.datetime(2012, 11, 10)), 

                	 (dt.datetime(2020, 11, 20), dt.datetime(2020, 12, 5)) 

                 ] 

  

dcsc_periods = [ 

                (dt.datetime(2011,12, 1), dt.datetime(2011, 12, 30)), 

                 (dt.datetime(2012, 10, 1), dt.datetime(2012, 12, 30)), 

(dt.datetime(2015, 11, 1), dt.datetime(2015, 11, 30)), 

(dt.datetime(2016, 5, 1), dt.datetime(2016, 5, 30)), 

(dt.datetime(2016, 11, 1), dt.datetime(2016, 11, 30)), 

(dt.datetime(2018, 11, 1), dt.datetime(2018,11, 30)), 

(dt.datetime(2020, 11, 1), dt.datetime(2020, 11, 30)), 

(dt.datetime(2021, 11, 1), dt.datetime(2021, 11, 30)), 

(dt.datetime(2022, 11, 1), dt.datetime(2022, 12, 30))] 

 
 
 

adt= pd.read_csv('average_adt_Tuticorin.csv')

# Convert the 'time' column to a datetime object
adt['time'] = pd.to_datetime(adt['time'])

# Set the 'time' column as the DataFrame index
adt.set_index('time', inplace=True)

# Filter data for time starting from 2011
adt = adt[adt.index >= '2011-01-01']
sl = adt['adt'] * 1000
sl_m = sl - sl.mean()


window_size = (3*24)+1
filtered_sl_anomaly = medfilt(df['sl_anomaly'], kernel_size=window_size)

# Create a new DataFrame to store the filtered data
filtered_df = df.copy()
filtered_df['filtered_sl_anomaly'] = filtered_sl_anomaly


mean_sl_anomaly = df['sl_anomaly'].mean()
st = df['sl_anomaly'].std()


df['sl_anomaly_filtered'] = df['sl_anomaly']
upper_limit=mean_sl_anomaly +(1.65*st)
lower_limit=mean_sl_anomaly -(1.65*st)
# Set values above the upper limit to NaN
df.loc[df['sl_anomaly'] > upper_limit, 'sl_anomaly_filtered'] = np.nan

# Set values below the lower limit to NaN
df.loc[df['sl_anomaly'] < lower_limit, 'sl_anomaly_filtered'] = np.nan


filtered_time = df['Time'].values[~np.isnan(df['sl_anomaly_filtered'])]
filtered_sl_anomaly = df['sl_anomaly_filtered'].values[~np.isnan(df['sl_anomaly_filtered'])]


# Interpolate missing values in 'sl_anomaly_filtered'
df['sl_anomaly_filtered_fill']=df['sl_anomaly_filtered'].interpolate(method='linear')
df['sl_anomaly_filtered_fill'].fillna(method='bfill', inplace=True)

time = df['Time'].values

# Create an array for x-values (time)
x = np.arange(len(time))

sl_anomaly_filtered_fill = df['sl_anomaly_filtered_fill'].values

# Perform linear regression to find the best-fitting line
slope, intercept, _, _, _ = linregress(x, sl_anomaly_filtered_fill)

# Calculate the trend line
trend_line = (slope * x )+ intercept
slope_mm_per_year=slope*365*24


plt.figure(figsize=(12, 6))

for dcsc_start, dcsc_end in dcsc_periods:
    plt.gca().axvspan(dcsc_start, dcsc_end, color='cyan', alpha=0.5)
for flood_start, flood_end in flood_periods:
    plt.gca().axvspan(flood_start, flood_end, color='orange', alpha=0.8)

flood_legend = mpatches.Patch(color='orange', alpha=0.8, label='Flood Period')
dcsc_legend = mpatches.Patch(color='cyan', alpha=0.5, label='D_C_SC Period')


ax = plt.gca()
plt.plot(df['Time'],  filtered_df['filtered_sl_anomaly'], linestyle='-', color='blue', linewidth=0.7, label='3day_med_fil_sla INCOIS')
plt.plot(df['Time'], trend_line, linestyle='--', color='black', linewidth=1.5,
         label='Trend line')
# plt.plot(df['Time'], linear_regression_line, linestyle='--', color='black', linewidth=2.5,
#          label=f'Linear Regression trend ({slope_mm_per_year:.2f} mm/year)')
plt.plot(adt.index, sl_m, color='red', linewidth=1.5, label='daily_sla_Satellite Altimetry')

# Set the background color of the axis to a custom color (e.g., light gray)
ax.set_facecolor('lightgray')

legend1 = plt.legend(handles=[flood_legend, dcsc_legend], loc='upper left')

# Add a legend for the trend lines
custom_lines = [mlines.Line2D([], [], color='blue', label='3day_med_fill_sla INCOIS'),
                mlines.Line2D([], [], color='red', label='daily_sla_Satellite Altimetry'),
                mlines.Line2D([], [], color='black',label=f'Trend line', linestyle='--')
               ]

legend2 = plt.legend(handles=custom_lines, loc='upper right')

# Add both legends to the plot
plt.gca().add_artist(legend1)
plt.gca().add_artist(legend2)
plt.title(f'Sea level trend for Tuticorin(Slope = {slope_mm_per_year:.2f} mm/yr)')
plt.xlabel('Time')
plt.ylabel('Sea level in mm')
plt.ylim(-1500, 1500)
plt.grid(True, linestyle=':', linewidth=0.8, color='Ivory', alpha=0.5)
plt.savefig(f'Trend_for_Tuticorin')
plt.show()
```
